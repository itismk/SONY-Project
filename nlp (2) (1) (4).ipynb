```python
import pandas as pd
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Load the saved model and vectorizer
with open('svm_model.pkl', 'rb') as f:
    clf, vectorizer = pickle.load(f)

# Load the new dataset
new_df = pd.read_csv('hurricane_harvey.csv')

# Preprocess the data
new_df['tweet_text'] = new_df['tweet_text'].apply(preprocess_text)

# Convert text data to numerical vectors using the same TF-IDF vectorizer used in training
X_new_vectors = vectorizer.transform(new_df['tweet_text'])

# Make predictions on the new dataset
y_pred = clf.predict(X_new_vectors)

# Create a new DataFrame consisting of the tweet_text and predicted text_info columns
new_df_pred = pd.DataFrame({'tweet_text': new_df['tweet_text'], 'predicted_text_info': y_pred})

# Apply K-means clustering to the numerical vectors
num_clusters = 7
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X_new_vectors)

# Add the cluster labels to the new DataFrame
new_df_pred['cluster'] = kmeans.labels_

# Visualize the clusters using a scatter plot
colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'brown', 'gray']
plt.figure(figsize=(8, 6))
for i in range(num_clusters):
    cluster_data = new_df_pred[new_df_pred['cluster'] == i]
    plt.scatter(cluster_data.index, np.zeros_like(cluster_data.index), color=colors[i], label='Cluster {}'.format(i+1))
plt.title('Clusters')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
#import nltk
#nltk.download('stopwords')
#from nltk.corpus import stopwords
import re


# Perform data cleaning and preprocessing
#stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    text = re.sub(r'http\S+', '', text) # Remove URLs
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # Remove @mentions
    text = re.sub(r'#', '', text) # Remove hashtags
    text = re.sub(r'[^\w\s]', '', text) # Remove punctuation
    text = text.lower() # Convert to lowercase
    #text = ' '.join([word for word in text.split() if word not in stop_words]) # Remove stopwords
    return text


# Create a list of file paths
file_paths = ['california_wildfires_final_data.csv', 'hurricane_maria.csv','srilanka.csv','hurricane_imra.csv','mexico.csv','californiamain.csv','iran.csv']

# Read the CSV files and concatenate them into a single DataFrame
df = pd.concat([pd.read_csv(f) for f in file_paths])

# Drop rows with null values in the tweet_text and text_info column
df.dropna(subset=['tweet_text'], inplace=True)
df.dropna(subset=['text_human'], inplace=True)

# Preprocess the data
df['tweet_text'] = df['tweet_text'].apply(preprocess_text)


# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['tweet_text'], df['text_human'], test_size=0.2, random_state=42)

# Extract features from the text using a TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)

# Define the hyperparameters to tune
params = {'C': [0.1, 1, 10], 'penalty': ['l1', 'l2']}

# Create a grid search object and fit it to the training data
grid = GridSearchCV(LogisticRegression(solver='liblinear'), params, cv=5)
grid.fit(X_train_features, y_train)

# Make predictions on the test set using the best model
best_model = grid.best_estimator_
y_pred = best_model.predict(X_test_features)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))


# Save the trained model as a .pkl file
with open('logistic_model.pkl', 'wb') as f:
    pickle.dump((vectorizer,  best_model), f)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


# Load the data
file_paths = ['california_wildfires_final_data.csv', 'hurricane_maria.csv','srilanka.csv','hurricane_imra.csv','mexico.csv','iran.csv']

# Read the CSV files and concatenate them into a single DataFrame
df = pd.concat([pd.read_csv(f) for f in file_paths])

train_data = df
test_data = pd.read_csv("hurricane_harvey.csv")

# Preprocess the data
le = LabelEncoder()
train_labels = le.fit_transform(train_data['tweet_text'])
test_labels = le.transform(test_data['tweet_text'])

tfidf = TfidfVectorizer()
train_features = tfidf.fit_transform(train_data['text_info'])
test_features = tfidf.transform(test_data['text_info'])

# Train the model
model = LogisticRegression()
model.fit(train_features, train_labels)

# Evaluate the model on the test data
y_pred_labels = model.predict(test_features)
accuracy = accuracy_score(test_labels, y_pred_labels)
print("Accuracy: {:.2f}%".format(accuracy*100))

# Use the model to predict on new data
new_data = pd.DataFrame({
    'text_info': ['This is a test', 'Fake news alert']
})

# Preprocess the new data
new_data_labels = le.transform(new_data['tweet_text'])
new_data_features = tfidf.transform(new_data['text_info'])

# Evaluate the model on the new data
y_pred_labels = model.predict(new_data_features)
accuracy = accuracy_score(new_data_labels, y_pred_labels)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import pickle 
import re

# Perform data cleaning and preprocessing
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove mentions
    text = re.sub(r'@\S+', '', text)
    # Remove hashtags
    text = re.sub(r'#\S+', '', text)
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Remove digits
    text = re.sub(r'\d+', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)
    # Return preprocessed text
    return text.strip()

# Create a list of file paths
file_paths = ['california_wildfires_final_data.csv', 'hurricane_maria.csv','srilanka.csv','hurricane_imra.csv','mexico.csv','iran.csv','californiamain.csv']

# Read the CSV files and concatenate them into a single DataFrame
df = pd.concat([pd.read_csv(f) for f in file_paths])

# Drop rows with null values in the tweet_text and text_info column
df.dropna(subset=['tweet_text'], inplace=True)
df.dropna(subset=['text_info'], inplace=True)

# Preprocess the data
df['tweet_text'] = df['tweet_text'].apply(preprocess_text)
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['tweet_text'], df['text_info'], test_size=0.2, random_state=42)

# Tokenize the text data
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X_train)
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad the sequences to have the same length
max_len = 100
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len)

# Build the RNN model
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=32, input_length=max_len))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(X_train_pad, y_train, epochs=10, batch_size=64)

# Evaluate the model
loss, accuracy = model.evaluate(X_test_pad, y_test)
print("Accuracy: {:.2f}%".format(accuracy*100))
with open('rnn_model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense
from tensorflow.keras.callbacks import EarlyStopping

# Load the data
file_paths = ['california_wildfires_final_data.csv', 'hurricane_maria.csv','srilanka.csv','hurricane_imra.csv','mexico.csv','iran.csv','californiamain.csv']

# Read the CSV files and concatenate them into a single DataFrame
df = pd.concat([pd.read_csv(f) for f in file_paths])
data = df
data = data[['tweet_text', 'text_info']]
data = data.rename(columns={'tweet_text': 'text_info', 'text_info': 'label'})

data['label'] = data['label'].map({'ham': 0, 'spam': 1})

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data['text_info'], data['label'], test_size=0.2, random_state=42)

# Tokenize the text
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(X_train)
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad the sequences
X_train_padded = pad_sequences(X_train_seq, maxlen=117036, padding='post')
X_test_padded = pad_sequences(X_test_seq, maxlen=117036, padding='post')

# Define the model
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=100, input_length=117036))
model.add(Conv1D(filters=128, kernel_size=5, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())

# Train the model
early_stop = EarlyStopping(monitor='val_loss', patience=3)
model.fit(X_train_padded, y_train, validation_split=0.2, epochs=10, callbacks=[early_stop])

# Evaluate the model
loss, accuracy = model.evaluate(X_test_padded, y_test)
print(f'Test loss: {loss:.3f}, accuracy: {accuracy:.3f}')


with open('CNNmodel.pkl', 'wb') as file:
    pickle.dump(model, file)
```
